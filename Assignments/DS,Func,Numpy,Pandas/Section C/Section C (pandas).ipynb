```python
import pandas as pd
import numpy as np

# Create a DataFrame manually
data = {
    "Name": ["Amit", "Riya", "Karan", "Sneha", "Amit", "Vijay", "Pooja", "Ravi", "Neha", "Riya"],
    "Age": [25, 22, 28, 24, 25, 30, 26, 29, 23, 22],
    "City": ["Delhi", "Mumbai", "Bangalore", "Chennai", "Delhi",
             "Hyderabad", "Pune", "Kolkata", "Jaipur", "Mumbai"],
    "Salary": [50000, 45000, 60000, np.nan, 50000,
               70000, 55000, np.nan, 48000, 45000]
}

df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)
df
```

```python
print("\nDataFrame Info:")
df.info()

print("\nDataFrame Description:")
print(df.describe())
```
