```python
# Create a list of 5 integers
numbers = [10, 20, 30, 40, 50]
print("Original list:", numbers)

# append()
numbers.append(60)
print("After append(60):", numbers)

# extend()
numbers.extend([70, 80])
print("After extend([70, 80]):", numbers)

# insert()
numbers.insert(2, 25)
print("After insert(2, 25):", numbers)

# remove()
numbers.remove(40)
print("After remove(40):", numbers)

# pop()
removed_element = numbers.pop()
print("After pop():", numbers)
print("Popped element:", removed_element)

# index()
index_of_30 = numbers.index(30)
print("Index of 30:", index_of_30)

# count()
count_of_20 = numbers.count(20)
print("Count of 20:", count_of_20)

# sort()
numbers.sort()
print("After sort():", numbers)

# reverse()
numbers.reverse()
print("After reverse():", numbers)

# clear()
numbers.clear()
print("After clear():", numbers)
```
